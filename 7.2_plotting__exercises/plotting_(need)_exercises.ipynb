```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
x=np.array(range(51))
x = np.linspace(0,51,200)
sinx=np.sin(x)
e=np.e
x1=e**np.divide(-x,10)
y=x1*sinx
plt.plot(x,y,color='blue')
plt.xlabel('x',fontsize='15',color='darkorange')
plt.ylabel('y',fontsize='15',color='darkorange',rotation=0)
plt.ylim([-1,1])
plt.title('$y=e^{-x/10}sin(x)$',fontsize='20',style='italic')
plt.grid(alpha=.25)
plt.show()
```
